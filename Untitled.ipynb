## Counting the types of atoms

```python
import collections

num_nodes_train = collections.defaultdict(int)
num_edges_train = collections.defaultdict(int)
node_types_train = collections.defaultdict(int)
node_degrees_train = collections.defaultdict(int)
edge_types_train = collections.defaultdict(int)

num_nodes_test = collections.defaultdict(int)
num_edges_test = collections.defaultdict(int)
node_types_test = collections.defaultdict(int)
node_degrees_test = collections.defaultdict(int)
edge_types_test = collections.defaultdict(int)

labels_count = collections.defaultdict(int)

for index in range(len(train)):
    G = train[index]
    n_nodes = len(G.nodes)
    n_edges = len(G.edges)
    
    for node in range(n_nodes):
        node_types_train[G.nodes[node]['labels'][0]] += 1
        node_degrees_train[G.degree[node]] += 1
        
        if len(G.nodes[node]['labels']) > 1:
            print(index)
            print(node)
            print(G.nodes[node]['labels'])
            raise ValueError
    
    for edge in G.edges:
        edge_types_train[G.edges[edge]['labels'][0]] += 1
        
        if len(G.edges[edge]['labels']) > 1:
            print(index)
            print(edge)
            print(G.edges[edge]['labels'])
            raise ValueError
            
    num_nodes_train[n_nodes] += 1
    num_edges_train[n_edges] += 1
    labels_count[labels[index]] += 1
    
for index in range(len(test)):
    G = train[index]
    n_nodes = len(G.nodes)
    n_edges = len(G.edges)
    
    for node in range(n_nodes):
        node_types_test[G.nodes[node]['labels'][0]] += 1
        node_degrees_test[G.degree[node]] += 1
        
        if len(G.nodes[node]['labels']) > 1:
            print(index)
            print(node)
            print(G.nodes[node]['labels'])
            raise ValueError
    
    for edge in G.edges:
        edge_types_test[G.edges[edge]['labels'][0]] += 1
        
        if len(G.edges[edge]['labels']) > 1:
            print(index)
            print(edge)
            print(G.edges[edge]['labels'])
            raise ValueError
            
    num_nodes_test[n_nodes] += 1
    num_edges_test[n_edges] += 1
```
